Importação das bibliotecas necessárias para à atividade

```python
import pandas as pd
import seaborn as srn
import statistics as sts
```

Carregamento do arquivo CSV

```python
df = pd.read_csv("tempo.csv", sep=";")
```

#Tratamento dos dados no arquivo tempo

Verificando a presença de valores NaN no data frame

```python
#Código para contar os valores NaN em todas as colunas
df.isnull().sum()

"""
Verificado a presença de valores NaN na coluna: Umidade.
Verificado a presença de valores NaN na coluna: Vento.
"""
```

Tratamento da coluna Aparencia

```python
#Agrupamento e visualização dos dados contidos na coluna Aparencia
agrupamento = df.groupby(['Aparencia']).size()
agrupamento

#Substituição do valor fora do padrão pelo valor da moda que é sol
df.loc[df['Aparencia'] == 'menos', 'Aparencia'] = 'sol'

#Verificando se o valor 'menos' realmente foi substituido corretamente
agrupamento = df.groupby(['Aparencia']).size()
agrupamento
```

Tratamento e substituição dos valores outliers na coluna Temperatura

```python
#Verificando a presença de valores outliers na coluna Temperatura
df['Temperatura'].describe()
"""Constatado o valor máximo de 1220 na coluna temperatura, ou seja, um valor outlier, ninguém sobrevive à essa temperatura."""

#Criação da variável que armazena o valor da mediana da coluna Temperatura
mediana = sts.median(df['Temperatura'])
mediana

#Substituição do valor de outlier pelo valor da variável denominada mediana
df.loc[df['Temperatura'] >= 2 * mediana, 'Temperatura'] = mediana

#Verificando novamente a descrição dos valores da coluna Temperatura
df['Temperatura'].describe()
```

Tratamento e verificação de valores Outliers na coluna Umidade

```python
#Verificando a presença de valores outliers na coluna Umidade
df['Umidade'].describe()

#Criação da variável que armazena o valor da mediana da coluna Umidade
mediana = sts.median(df['Umidade'])
#Criação da variável que armazena o valor do desvio padrão da coluna Umidade
desvio = df['Umidade'].std()

#Preenchimento dos valores NaN pela mediana
df['Umidade'] = df['Umidade'].fillna(mediana)

#Checando se ainda existem valores NaN na coluna Umidade
df.isnull().sum()
"""Não há mais valores NaN na coluna Umidade"""

#Substituição do valor Outlier na coluna Umidade
df.loc[df['Umidade'] >= 100, 'Umidade'] = mediana

#Verificando a substituição de valores outliers
df['Umidade'].describe()
"""Valor outlier substituído pela mediana"""
```

Tratamento da coluna Vento

```python
#Verificação da coluna vento em busca do valor da moda
agrupamento = df.groupby(['Vento']).size()
agrupamento
#Verificado que o valor da moda da coluna Vento é FALSO

#Substituição do valor NaN pelo valor da moda
df['Vento'] = df['Vento'].fillna('FALSO')

#Verificando se o valor NaN foi substituido corretamente
agrupamento = df.groupby(['Vento']).size()
agrupamento
```

Exportação dos dados tratados para um arquivo CSV

```python
#Exportando o tratamento dos dados para um arquivo CSV
df2 = df.to_csv('tempo_tratado.csv', sep=";")
df2
```
