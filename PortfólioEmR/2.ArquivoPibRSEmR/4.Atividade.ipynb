Checando se existem duplicadas no dataframe

```r
x = df[duplicated(df$CODIGO), ]
x
"Dados que estão duplicados, vamos excluir pelo código das cidades"
#259, 141, 378, 380, 372
```

```r
#Excluindo os dados duplicados pelo código
df = df[!df$CODIGO %in% c(259, 141, 378, 380, 372), ]
```

```r
#Verificando novamente se ainda existem dados duplicados
df[df$CODIGO == x$CODIGO, ]
```

```r
#Verificamos novamente duplicados
x = df[duplicated(df$CODIGO), ]
x
```

Verificando se existem valores NAs na tabela

```r
df[!complete.cases(df), ]
"Constatado que não há presença de valores NAs na tabela dados"
```

#Visualizando dados estatísticos da coluna PIB

Visualizando o summary novamente do data frame

```r
summary(df$VALOREMPENHO)
```

```r
summary(df$PIB)
```

Busca os maiores empenhos

```r
maiorempenho = head(df[order(-df$VALOREMPENHO),],10)
barplot(maiorempenho$VALOREMPENHO, col=brewer.pal(n = 10, name = "RdBu"),las=2,main = "Maiores Empenhos")
legend("topright",legend=maiorempenho$MUNICIPIO,col = brewer.pal(n = 10, name = "RdBu"), lty=1:2, cex=0.8, ncol = 2,lwd=4)
box()
```

Maiores Pibs

```r
maiorpib =   head(df[order(-df$PIB),],10)
barplot(maiorpib$PIB,col=brewer.pal(n = 10, name = "Spectral"),las=2,main = "Maiores PIBS")
legend("topright",legend=maiorempenho$MUNICIPIO,col = brewer.pal(n = 10, name = "Spectral"), lty=1:2, cex=0.8, ncol = 2,lwd=4)
box()
```

Calcula qual % de PIB foi empenhado

```r
df$PROPORCAO =   df$VALOREMPENHO  / df$PIB
Mprop = head(df[order(-df$PROPORCAO),],10)
barplot(Mprop$PROPORCAO,col=brewer.pal(n = 10, name = "Set2"),las=2,main = "Maiores Gastos em Proporção ao PIB")
legend("topright",legend=Mprop$MUNICIPIO,col = brewer.pal(n = 10, name = "Set2"), lty=1:2, cex=0.8, ncol = 2,lwd=4)
box()
```
